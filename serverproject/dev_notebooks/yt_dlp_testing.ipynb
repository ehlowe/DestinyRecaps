```python
import subprocess

folder_path = "destinyapp/working_folder/working_audio/"

ydl_opts = {
    'format': 'bestaudio/best',
    'outtmpl': folder_path + 'raw' + '.%(ext)s',
    'age_limit': 21,
    'username': 'oauth2',
    'password': '',
    'verbose': True,
}

video_id="H-XdAZC9U7M"

# Construct the yt-dlp command
url = "https://www.youtube.com/watch?v="+video_id  # Replace with your desired URL
command = [
    "yt-dlp",
    "--username", "oauth2",
    "--password", "",
    "--extract-audio",
    "-o", ydl_opts['outtmpl'],
    "--age-limit", str(ydl_opts['age_limit']),
    "--verbose" if ydl_opts['verbose'] else "",
    url,
]


command = [arg for arg in command if arg]

# Execute the command
try:
    result = subprocess.run(command, check=True, capture_output=True, text=True)
    print("Command output:", result.stdout)
except subprocess.CalledProcessError as e:
    print("An error occurred:", e)
    print("Error output:", e.stderr)
```
